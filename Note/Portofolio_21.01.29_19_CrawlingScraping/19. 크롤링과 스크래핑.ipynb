# 데이터 다운로드 하기 -- 02
- PC에 바로가져오는 방식과는 다르다
- 인터넷에서 지정된 파일을 메모리에 올리고 저장하기


```python
import urllib.request

# URL과 저장경로 지정하기.
url = "http://uta.pw/shodou/img/28/214.png"
savename = "../Data/test02.png"

# 다운로드 하기
# 작업(색깔,..등등을 변경)하기 위해 
men = urllib.request.urlopen(url).read()

# 파일로 저장하기 
# wb = write binary의 약어.
with open(savename, "wb") as f:
    f.write(men)
    
print("저장 되었습니다!.")
```

# 매개변수를 추가해 요청을 전송하는 방법
- Site 이다
- xml이다.
-기상청 RSS : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp         
    - URL 뒤에 매개변수를 입력하여 지역별로 일기 정보를 가져올 수 있음.     
      - 전국 108, 서울 / 경기도 : 109, 강원도 : 105, 제주특별자치도 : 184         
ex) 서울/경기도 
            - http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109

```python
import urllib.request
import urllib.parse # parsing 할때 

API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 URL Encoding 해야 볼수 있다!!
# stnId-> Dictionary로 만들어야함.
values = {'stnId' : '109'}
params = urllib.parse.urlencode(values)

# 요청 전용 URL 생성
url = API + "?" + params
print("url = ", url )

# 데이터 가져오기

data = urllib.request.urlopen(url).read()
# print(data) # 글자를 확인할 수 없음.
# 그래서 decode가 필요 하다.
text = data.decode('utf-8')
print(text)

```

# BeautifulSoup로 Scraping 하기
- BeautifulSoup : 간단하게 HTML과 XML에서 정보 추출
- Beautiful 설치 : 터미널에서 pip install beautifulsoup4

```python
# 기본 사용법
from bs4 import BeautifulSoup

# HTML Sample
# -> 이거 끝나고 HTML SITE에 들어가서 가져오면 됨
html = """
        <html>
            <body>
                <h1> Header </h1>
                <p> Line 1을 서술 </p>
            </body>
        </html>


"""

# HTML 분석
# html.parser 방식으로 분리를 해라 의미
soup = BeautifulSoup(html, 'html.parser')
#print(soup) - 잘 들어 왔는지 확인.

# 원하는 부분 추출하기
h1 = soup.html.body.h1 
#print(h1) - 잘 가져 왔는지 확인.
p = soup.html.body.p
#print(p)

# text만 출력 할거야!
print("h1 =", h1.string)
print("p =", p.string)

```

```python
# id로 요소를 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <h1 id ="title"> Header </h1>
                <p id = "body"> Line 1을 서술 </p>
            </body>
        </html>


"""

# HTML 분석
# html.parser 방식으로 분리를 해라 의미
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
title = soup.find(id ='title')
body = soup.find(id='body')
print(title)
print(body)

# text만 출력하기
print('title =', title.string)
print('body =', body.string)

# 어쩔때 .string이 안먹히는 경우 -> 이모티콘, 폰트가 다르게 있을경우 안될수 있다
# 그럴떄 .text를 쓴다.
print('body =', body.text)

```

```python
# 여러개의 요소 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
            <ul>
                <li><a href="http://www.naver.com">naver</a></li>
                <li><a href="http://www.daum.net">daum</a></li>
            </ul>
            </body>
        </html>

"""

# HTML 분석
# html.parser 방식으로 분리를 해라 의미
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
links = soup.find_all('a')
print(links) # list에 담겨서 옴. 

for link in links:
    # href 태그값 가져오기
    href = link.attrs['href']
    # naver라는 이름 가져오기
    text = link.string
    # 확인
    print(text, "==>", href)
```

# CSS 선택자 사용하기
- soup.select_one() : CSS 선택자로 요소 하나를 추출
- soup.select() : CSS 선택자로 여러개를 리스트로 추출\

```python
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <div id = "meigen">
                    <h1>위키 북스</h1>
                    <ul class ="items">
                       <li>유니티 게임 이팩트 입문서</li>
                       <li>스위프로 시작하는 아이폰 앱 개발 교과서</li>
                       <li>모던 웹사이트 디자인의 정석</li>
                    </ul>
                </div>
            </body>
        </html>
"""

# HTML 분석 하기
# CSS 든 XML이든 html.parser이다.
soup = BeautifulSoup(html, "html.parser")

# 필요한 부분 CSS로 추출하기
# Title 부분 추출 하기 
# div#ID값 적으면 된다.
# id: #, class: . , tag: tag명
# 자손: >, 후손: ' ' 


h1 = soup.select_one('div#meigen > h1').string
print("h1:",h1)

# 목록 부분 추출하기
#li_list=soup.select("div#meigen > ul.items > li") # 안먹힐 경우 위치를 정확하게 알려준다.
#for li in li_list:
#    print("li=",li.string)
li_lists =soup.select("ul.items > li")

for li in li_lists: 
    print('li :', li.string)
```

# Rotten Tomatoes
https://www.rottentomatoes.com/top/bestofrt/?year=2019

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
```

```python
# HTML 가져오기 : 다운로드 받은 후 사용
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)
# HTML 분석하기
soup = BeautifulSoup(res,"html.parser")
```

```python
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
soup.select("#top_movies_main > div > table > tr > td > a ")[0].string
```

# 적용

```python
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
li_lists = soup.select("#top_movies_main > div > table > tr > td > a ")
# 100인지 아닌지 볼려면 
count = 0 
titleLists = [] # 영화제목 들어갈 Lists
countLists = [] # Count Lists

for li in li_lists:
    count += 1
    print(count, li.string.strip()) # strip()띄워쓰기 앞뒤로 없애기
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countLists만으로 dataframe 생성
df = pd.DataFrame(data = countLists, columns=['index'])
df
```

```python
df['Titles'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
